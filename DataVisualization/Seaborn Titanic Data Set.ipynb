```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style('whitegrid')
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

## Exercise

```python
sns.jointplot(x='fare', y='age', data=titanic, color='blue')
```

```python
sns.distplot(titanic['fare'], kde=False, color='red')
```

```python
sns.distplot(titanic['fare'], kde=False, color='red', bins=30)
```

```python
sns.boxplot(x='class', y='age', data=titanic, palette='rainbow')
```

```python
sns.swarmplot(x='class', y='age', data=titanic, palette='Set1')
```

```python
sns.countplot(x='sex', data=titanic)
```
