```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y =  x**2
```

```python
x
```

```python
y
```

```python
# FUNCTIONAL
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1) #n of rows,columns, plot number
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# OBJECT-ORIENTED METHOD
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 =fig.add_axes([0.1,0.1,0.8,0.8])
axes2 =fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('LARGER PLOT')

axes2.plot(y,x)
axes2.set_title('SMALLER PLOT')
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)

#axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)

for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x,y)
axes[0].set_title("First")

axes[1].plot(y,x)
axes[1].set_title("Second")

plt.tight_layout() #takes care of overlapping plots
```

```python
fig = plt.figure(figsize=(3,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig = plt.figure(figsize=(6,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1,figsize=(5,2))

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig.savefig('my_pic.png', dpi=200)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label = 'X Squared')
ax.plot(x, x**3, label = 'X Cubed')

ax.legend(loc=0) #loc for legend location
```

Part3: Plot Appearance

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, color='#FF8C00', linewidth=2, linestyle='-.', marker='o', markersize=6,
       markerfacecolor='blue', markeredgewidth=3, markeredgecolor='green')

#alpha = transparency
#linewidth = lw
#linestyle= ls
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, color='blue', linewidth=2, linestyle='-.')

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
