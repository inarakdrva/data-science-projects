```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('lenna.png')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap="gray")
```

Task 1. Примените операцию binary thresholding на изображение gray.

```python
img = cv2.imread('lenna.png', cv2.IMREAD_GRAYSCALE)
_, binary_image = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 2)
plt.imshow(binary_image, cmap='gray')

plt.show()
```

Task 2. Примените любой blurring operation.

```python
blurred_image = cv2.GaussianBlur(img, (15, 15), 0)

plt.figure(figsize=(10,5))
plt.subplot(1, 2, 2)
plt.imshow(blurred_image, cmap='gray')

plt.show()
```

Task 3: Разверните изображение справа налево

```python
flipped_img = cv2.flip(img, 1)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(flipped_img, cv2.COLOR_BGR2RGB))

plt.show()
```

Task 4: Разверните изображение сверху вниз И справа-налево

```python
flipped_img_horizontal = cv2.flip(img, 1)

flipped_img_both = cv2.flip(flipped_img_horizontal, 0)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(flipped_img_both, cv2.COLOR_BGR2RGB))

plt.show()
```

Task 5. Обрежьте и оставьте только лицо с изображения. Добавьте к нему текст «Я Ленна»

```python
image_crop = img[210:390,220:370]

text = "I am Lenna"
font = cv2.FONT_HERSHEY_SIMPLEX
font_scale = 0.8
font_thickness = 2
text_size = cv2.getTextSize(text, font, font_scale, font_thickness)[0]
text_x = (image_crop.shape[1] - text_size[0]) // 2
text_y = (image_crop.shape[0] + text_size[1]) // 2
cv2.putText(image_crop, text, (text_x, text_y), font, font_scale, (255, 255, 255), font_thickness)

plt.imshow(cv2.cvtColor(image_crop, cv2.COLOR_BGR2RGB))
plt.show()
```

Task 7. Используя bitwise операторы, постройте следующию фигуру, используя прямоугольник и круг.

```python
canvas = np.zeros((300, 300), dtype=np.uint8)

rectangle = cv2.rectangle(canvas.copy(), (25, 25), (275, 275), 255, -1)
circle = cv2.circle(canvas.copy(), (150, 150), 150, 255, -1)

result = cv2.bitwise_or(rectangle, circle)

plt.imshow(result, cmap='gray')
plt.show()
```
